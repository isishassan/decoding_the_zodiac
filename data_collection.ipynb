## Data Cleaning

- hindustan times: remove everything starting with "love focus" (incl)
- replace all sign names with [Sign Name]

```python
#replace sign names

combined["horoscope"] = combined["horoscope"].replace(
    ["Aries","Taurus","Gemini","Cancer","Leo","Virgo","Libra","Scorpio","Sagitarrius","Capricorn","Aquarius","Pisces"],
    "[Sign Name]"
)
```

```python
import pandas as pd

# Show full column content
pd.set_option('display.max_colwidth', None)
```

```python
filtered_df = combined[combined['horoscope'].str.contains('Aries', case=False, na=False)]

filtered_df 
```

```python
combined['horoscope'] = combined['horoscope'].replace(
    to_replace=r'Aries|Taurus|Gemini|Cancer|Leo|Virgo|Libra|Scorpio|Sagittarius|Capricorn|Aquarius|Pisces',  # regex pattern matching either word
    value='[Sign Name]',
    regex=True
)
```

```python
#cutting out unnecessary stuff from hindustan times

#filtered_df = combined['source'] == "Hindustan Times"

filtered_df = combined[combined['source'] == 'Hindustan Times']
display(filtered_df)
```

```python
#removing everything related to lucky number and colour
combined['horoscope'] = combined['horoscope'].str.split('\n', n=1).str[0]
```

```python
#checking if it worked:
filtered_df = combined[combined['source'] == 'Hindustan Times']
display(filtered_df)
```

```python
#duplicates

combined.duplicated().sum()
```

```python
#dropping my dupe
combined_clean = pd.DataFrame.drop_duplicates(combined)
```
